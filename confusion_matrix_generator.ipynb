```python
import pandas as pd
import numpy as np

def create_confusion_matrix(df, title, states_order=['ACQ1', 'ACQ2', 'CAP1', 'CAP2', 'NEGO', 'SQ', 'WAR'], fig_title = ""):
    """
    Creates a confusion matrix from a DataFrame containing Prediction and GroundTruth columns.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame with columns 'Prediction' and 'GroundTruth'
    states_order : list
        Order of states for the confusion matrix axes
        
    Returns:
    --------
    confusion_matrix : pandas.DataFrame
        Confusion matrix with GroundTruth as rows (y-axis) and Predictions as columns (x-axis)
    metrics : dict
        Dictionary containing per-class and overall metrics
    """
    
    # Initialize confusion matrix
    confusion_matrix = pd.DataFrame(
        data=0, 
        index=states_order, 
        columns=states_order
    )
    
    # Populate confusion matrix
    for _, row in df.iterrows():
        pred = row['Prediction']
        truth = row['GroundTruth']
        if pred in states_order and truth in states_order:
            confusion_matrix.loc[truth, pred] += 1
    
    # Calculate metrics
    metrics = {}
    
    # Per-class metrics
    for state in states_order:
        tp = confusion_matrix.loc[state, state]
        
        # Total predicted for this class (column sum)
        total_predicted = confusion_matrix[state].sum()
        
        # Total actual for this class (row sum)
        total_actual = confusion_matrix.loc[state].sum()
        
        precision = tp / total_predicted if total_predicted > 0 else 0
        recall = tp / total_actual if total_actual > 0 else 0
        f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
        
        metrics[state] = {
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'support': total_actual,
            'true_positives': tp,
            'total_predicted': total_predicted
        }
    
    # Overall accuracy
    correct = sum(confusion_matrix.loc[state, state] for state in states_order)
    total = confusion_matrix.sum().sum()
    metrics['overall'] = {
        'accuracy': correct / total if total > 0 else 0,
        'correct': correct,
        'total': total
    }

    print(f"Overall Accuracy: {metrics['overall']['accuracy']:.2f}, Correct: {metrics['overall']['correct']}, Total: {metrics['overall']['total']}")
    

    plt.figure(figsize=(8, 6))
    sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues',
                xticklabels=labels, yticklabels=labels, cbar=True)

    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()

    plt.savefig(fig_title, dpi=300, bbox_inches='tight')

    plt.show()


```

```python
import pandas as pd
import numpy as np
from collections import Counter

# DFA - original daaset
# DFB - ued for parameter tuning
def create_test_dataset(DFA, DFB, samples_per_class=15, random_state=515):
    
    np.random.seed(random_state)
    
    print(f"Original DFA shape: {DFA.shape}")
    print(f"Original DFB shape: {DFB.shape}")
    print(f"Target: {samples_per_class} samples per class")
    
    # Create a composite key to identify unique dyad combinations
    DFA_key = DFA[['ccode1', 'ccode2']].apply(lambda x: f"{x['ccode1']}_{x['ccode2']}", axis=1)
    DFB_key = DFB[['ccode1', 'ccode2']].apply(lambda x: f"{x['ccode1']}_{x['ccode2']}", axis=1)
    
    # Find datapoints in DFA that are NOT in DFB
    mask_not_in_dfb = ~DFA_key.isin(DFB_key)
    DFA_filtered = DFA[mask_not_in_dfb].copy()
    
    print(f"\nAfter excluding DFB datapoints from DFA: {DFA_filtered.shape}")
    print(f"Excluded {DFA.shape[0] - DFA_filtered.shape[0]} datapoints that were in DFB")
    
    # Check groundtruth distribution in filtered data
    groundtruth_counts = DFA_filtered['groundtruth'].value_counts()
    print(f"\nGroundtruth distribution in filtered DFA:")
    for gt in ['ACQ1', 'ACQ2', 'CAP1', 'CAP2', 'NEGO', 'SQ', 'WAR1', 'WAR2']:
        count = groundtruth_counts.get(gt, 0)
        print(f"  {gt}: {count}")
    
    # Define all possible groundtruth values
    all_groundtruths = ['ACQ1', 'ACQ2', 'CAP1', 'CAP2', 'NEGO', 'SQ', 'WAR1', 'WAR2']
    
    # Sample from each groundtruth category
    sampled_dfs = []
    sampling_stats = {}
    
    for i, gt in enumerate(all_groundtruths):
        gt_data_filtered = DFA_filtered[DFA_filtered['groundtruth'] == gt]
        available_filtered_count = len(gt_data_filtered)
        
        if available_filtered_count >= samples_per_class:
            # Sufficient data available in filtered dataset
            sampled_data = gt_data_filtered.sample(n=samples_per_class, random_state=random_state + i)
            print(f"  {gt}: Sampled {samples_per_class} from {available_filtered_count} filtered datapoints")
            sampling_stats[gt] = {
                'filtered_available': available_filtered_count,
                'sampled': samples_per_class,
                'source': 'filtered_data'
            }
        
        elif available_filtered_count > 0:
            # Some data available in filtered dataset, but not enough
            # Use all filtered data and supplement from original DFA if possible
            remaining_needed = samples_per_class - available_filtered_count
            gt_data_original = DFA[DFA['groundtruth'] == gt]
            
            # Get additional samples from original DFA (excluding what we already took)
            additional_candidates = gt_data_original[~gt_data_original.index.isin(gt_data_filtered.index)]
            
            if len(additional_candidates) >= remaining_needed:
                additional_samples = additional_candidates.sample(n=remaining_needed, random_state=random_state + i + 100)
                sampled_data = pd.concat([gt_data_filtered, additional_samples], ignore_index=True)
                print(f"  {gt}: Used all {available_filtered_count} filtered + {remaining_needed} from original DFA")
                sampling_stats[gt] = {
                    'filtered_available': available_filtered_count,
                    'sampled': samples_per_class,
                    'source': 'mixed_filtered_and_original'
                }
            else:
                # Use all available data from both filtered and original
                all_available = pd.concat([gt_data_filtered, additional_candidates], ignore_index=True)
                sampled_data = all_available
                print(f"  {gt}: Used all available {len(sampled_data)} datapoints ({available_filtered_count} filtered + {len(additional_candidates)} original)")
                sampling_stats[gt] = {
                    'filtered_available': available_filtered_count,
                    'sampled': len(sampled_data),
                    'source': 'all_available'
                }
        
        else:
            # No data available in filtered dataset, sample from original DFA
            gt_data_original = DFA[DFA['groundtruth'] == gt]
            original_available_count = len(gt_data_original)
            
            if original_available_count >= samples_per_class:
                sampled_data = gt_data_original.sample(n=samples_per_class, random_state=random_state + i + 200)
                print(f"  {gt}: No filtered data available, sampled {samples_per_class} from {original_available_count} original datapoints")
                sampling_stats[gt] = {
                    'filtered_available': 0,
                    'sampled': samples_per_class,
                    'source': 'original_data_only'
                }
            else:
                # Use all available data from original DFA
                sampled_data = gt_data_original.copy()
                print(f"  {gt}: No filtered data available, used all {original_available_count} original datapoints")
                sampling_stats[gt] = {
                    'filtered_available': 0,
                    'sampled': original_available_count,
                    'source': 'all_original_available'
                }
        
        if len(sampled_data) > 0:
            sampled_dfs.append(sampled_data)
    
    # Combine all sampled data
    if sampled_dfs:
        DF_test = pd.concat(sampled_dfs, ignore_index=True)
        
        # Shuffle the final dataset
        DF_test = DF_test.sample(frac=1, random_state=random_state).reset_index(drop=True)
        
        print(f"\nFinal test dataset shape: {DF_test.shape}")
        print(f"Final groundtruth distribution:")
        final_counts = DF_test['groundtruth'].value_counts().sort_index()
        for gt in all_groundtruths:
            count = final_counts.get(gt, 0)
            print(f"  {gt}: {count}")
        
    else:
        DF_test = pd.DataFrame()
        print("No data available for sampling!")
    
    # Prepare exclusion statistics
    exclusion_stats = {
        'original_dfa_size': len(DFA),
        'original_dfb_size': len(DFB),
        'filtered_dfa_size': len(DFA_filtered),
        'excluded_count': len(DFA) - len(DFA_filtered),
        'final_test_size': len(DF_test),
        'sampling_stats': sampling_stats,
        'samples_per_class': samples_per_class,
        'target_total_size': samples_per_class * len(all_groundtruths)
    }
    
    return DF_test, exclusion_stats

```

```python
# Load your dataframes
DFA = groundtruth_df
DFB = train_df

# Create balanced test dataset
DF_test, stats = create_test_dataset(DFA, DFB, samples_per_class=15, random_state=42)

# Save the test dataset
DF_test.to_csv('test_balanced.csv', index=False)

# Display statistics
print("\n" + "="*50)
print("EXCLUSION AND SAMPLING STATISTICS")
print("="*50)
print(f"Original Dataset size: {stats['original_dfa_size']}")
print(f"Training Balanced Dataset size: {stats['original_dfb_size']}")
print(f"Filtered Original dataset size: {stats['filtered_dfa_size']}")
print(f"Excluded from Original dataset: {stats['excluded_count']}")
print(f"Final test size: {stats['final_test_size']}")
```
